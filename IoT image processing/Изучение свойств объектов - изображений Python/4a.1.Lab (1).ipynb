#### Шаг 9.  Добавление шума к объекту-изображению
Используя приводимую ниже функцию Pepper (или любые другие средства) добавьте случайный шум к объекту-изображению 
Отобразите нормализованный объект во вьюере ячейки.
<li>**Параметры функции pepper:**</li>
    
    image - объект
    rate - частота шума
    color - "цвет" шума

```python
def pepper (image,rate,color): #функция для добавления шума к изображению
    from random import uniform
    a,b,c=image.shape
    image_pep = image + 0
    for i in range(a):
        for j in range(b):
            if (uniform (0,1) > 1-rate) : 
                image_pep[i][j] = [color,color,color,255]
    return (image_pep)
```

```python
#Ячейка №9
canvas_pep = pepper(canvas,0.25,0)
canvas_pep.view()
plt.imshow(canvas_pep)
plt.show() #на выходе получаем зашумленное изображение
```

#### Шаг 10.  Устранение шума из объекта-изображения
Используя собственную функцию или известные вам стандартные средства Python устраните случайный шум из объекта-изображения. Отобразите нормализованный объект во вьюере ячейки.

```python
def unpepper (image,rate,color): #функция для устранения шума из зашумленного объекта
    a,b,c=image.shape
    image_pep=image+0
    for i in range (a-1):
        for j in range (b-1):
            if image[i][j][0]==image[i][j][1]==image[i][j][2]==color:
                image_pep[i][j]=image[i+1][j]
    return (image_pep)
#########Возможно для последующих курсов было бы более понятно как выолнять последнее задание, если бы в 9 задание был бы
#приведен помимо кода готовой функции также плюсом код объяснения, или лучше было бы не давать полностью готовый код,
# а давать лишь шаблоны заготовки с подробным объеснением, что делать дальше, а дальше реализация оставалась бы на стороне
# студента. На мой взгляд это бы улучшило бы восприятие материала (Лично мне подход в практике 4a2 понравился гораздо больше,
# т.к нет соблазна использовать готовый код, а искать решение и добывать информацию из документации python)
```
