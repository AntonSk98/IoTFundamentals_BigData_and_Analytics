
#### Шаг 6. Определите функцию для преобразования изображения по методу convolution (псевдо-свертка)

Определите функцию, которая формирует выходное изображение (Output) выполняя двумерную свертку входного изображения (input) с ядром (Kernel) 3x3 пиксела по приведенной ниже схеме: 
<img src="convolve-s.jpg">


```python
#Ячейка №6
def quonvolve2d(image, kernel):
    # This function which takes an image and a kernel 
    # and returns the convolution of them
    # Args:
    #   image: a numpy array of size [image_height, image_width].
    #   kernel: a numpy array of size [kernel_height, kernel_width].
    # Returns:
    #   a numpy array of size [image_height, image_width] (convolution output).
    
    kernel = np.flipud(np.fliplr(kernel))    # Flip the kernel
    output = np.zeros_like(image)            # convolution output
# Add zero padding to the input image
    image_padded = np.zeros((image.shape[0] + 2, image.shape[1] + 2))   
    image_padded[1:-1, 1:-1] = image.copy()
# Image normalization 
    for x in range(image.shape[1]):     # Loop over every pixel of the image
        for y in range(image.shape[0]):
            if (image_padded[y,x]> 240): image_padded[y,x]= 255  
# Image conversion                 
  # Loop over every pixel of the image
    for x in range(image.shape[1]):     # Loop over every pixel of the image
        for y in range(image.shape[0]):
            # element-wise multiplication of the kernel and the image
            output[y,x]=((kernel*image_padded[y:y+3,x:x+3]).sum())
    return output
```

```python
#Ячейка №7
kernel = np.array([[0,0,0],[0,1,0],[0,0,0]])
image_sharpen = quonvolve2d(gray,kernel)
plt.imshow(image_sharpen, cmap='gray')
plt.show()
viewer = ImageViewer(image_sharpen)
viewer.show()
```

Фильтры сглаживания (низкочастотные) сглаживают данные, уменьшая локальное изменение и удаляя шум. Низкочастотный фильтр вычисляет среднее значение для каждой окрестности. Эффект в том, что большие и маленькие значения в каждой окрестности будут усредняться, что уменьшит экстремальные значения данных.

```python
#тип сглаживания - сгладить среднеее арифметическое
kernel = np.array([[0.111,0.111,0.111],[0.111,0.111,0.111],[0.111,0.111,0.111]])
image_sharpen = quonvolve2d(myGray,kernel)
plt.imshow(image_sharpen, cmap='gray' )
plt.show()
viewer = ImageViewer(image_sharpen)
viewer.show()
```

```python
#фильтр ЛАПЛАСА
kernel = np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])
image_sharpen = quonvolve2d(myGray,kernel)
plt.imshow(image_sharpen, cmap='gray' )
plt.show()
viewer = ImageViewer(image_sharpen)
viewer.show()
```

ФИЛЬТР ЛАПЛАСА
