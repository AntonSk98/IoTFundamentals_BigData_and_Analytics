#### Шаг 3: Обучение и оценка модели дерева решений.

##### a) Создайте массив с переменной, которая будет являться целью для модели. 
Целью модели является классификация пассажиров на оставшихся в живых или нет. В процессе обучения модель будет учиться определять, какие значения входных переменных, скорее всего, указывают на принадлежность пассажира к жертвам или выжившим, а затем использовать эту информацию для классификации пассажиров из уникального набора тестовых данных.

````python
#Ячейка для кода №8
#создать массив для выходных значений
y_target = training["Survived"].values
````

##### b) Создайте массив значений, которые будут направленны на вход модели. 
Только некоторые характеристики данных полезны для создания дерева классификаторов. Мы создадим список столбцов из данных, которые мы хотим, чтобы классификатор использовал в качестве входных переменных, а затем создадим массив, используя названия столбцов из этой переменной. Переменная `X_input` содержит значения для всех характеристик, которые будут использоваться моделью для классификации. После обучения модели мы будем использовать эту переменную, чтобы присвоить эти метки набору тестовых данных.

````python
#Ячейка для кода №9
columns = ["Fare", "Pclass", "Gender", "Age", "SibSp"]
#создайте переменную для хранения характеристик, которые будет использовать классификатор
X_input = training[list(columns)].values
````

##### c) Создайте обученную модель.
Импортируйте модуль дерева решений из библиотеки машинного обучения `sklearn`. Создайте объект-классификатор ```clf_train```. Затем используйте метод ```fit()``` объекта классификатора, с помощью переменных X_input и y_target в качестве параметров, для обучения модели.

````python
#Ячейка для кода №10
#импортируйте модуль дерева из библиотеки sklearn
from sklearn import tree

#создайте clf_train как объект классификатора дерева решений
clf_train = tree.DecisionTreeClassifier(criterion="entropy", max_depth=3)

#обучите модель с помощью метода fit(). 
#Отправьте в метод входную переменную X_input и выходную переменную y_target
clf_train = clf_train.fit(X_input, y_target)
````

##### d) Оцените модель 
Используйте метод ```score()``` объекта дерева решений, чтобы отобразить в процентах точность присвоений, сделанных классификатором. В качестве аргументов он принимает входные и выходные переменные.

````python
#Ячейка для кода №11
clf_train.score(X_input,y_target)
````

Эта оценка показывает, что классификации, сделанные моделью, должны быть правильными примерно в 82% случаев.

Этот метод автоматически разбивает весь набор данных, возвращая в общей сложности четыре массива numpy, два для характеристик (тестирование и проверка) и два для меток (тестирование и проверка). Один параметр метода определяет долю наблюдений, используемых для тестирования и обучения. Другой параметр указывает начальное значение, которое будет использоваться для случайного назначения наблюдения на тестирование или обучение. Это используется для того, чтобы другой пользователь мог повторить вашу работу, получая те же самые назначения наблюдений наборам данных. Синтаксис метода:

<center>```train_test_split(input_X, target_y, test_size=0.4, random_state=0)```</center>

40% данных будут использоваться для тестирования. Случайное начальное значение устанавливается равным 0.


Метод возвращает четыре значения. Они возвращаются в следующем порядке: входные переменные для данных обучения, входные переменные для тестируемых данных, выходные переменные для данных обучения и выходные переменные для тестируемых данных.
